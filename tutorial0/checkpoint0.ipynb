```python
# fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 30 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number

for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial

import pandas as pd

starbucks_df = pd.read_csv("../data/starbucks.csv")

starbucks_df.head()


```

```python
# select all rows with more than and including 400 calories

calorie_above_400 = starbucks_df[starbucks_df["calories"] >= 400]
calorie_above_400
```

```python
# select all rows whose vitamin c content is higher than the iron content

vitamin_c_higher = starbucks_df[starbucks_df["vitamin a"] > starbucks_df["iron"]]
vitamin_c_higher
```

```python
# create a new column containing the caffeine per calories of each drink

starbucks_df["caffeine_per_calories"] = starbucks_df["caffeine"] / starbucks_df["calories"]

starbucks_df.columns
```

```python
# what is the average calorie across all items?

average_calories = starbucks_df["calories"].mean()

average_calories
```

```python
# how many different categories of beverages are there?

from collections import Counter

beverage_categories = Counter(starbucks_df["beverage_category"])

len(beverage_categories)
```

```python
# what is the average # calories for each beverage category?

average_calories_per_category = starbucks_df.groupby("beverage_category")["calories"].mean()

average_calories_per_category
```
